```python
import numpy as np
import matplotlib.pyplot as plt
import math
import numpy as np
from scipy.stats import entropy
import pandas as pd



def Ent(Data):
    pd_series = pd. Series(Data)
    counts = pd_series. value_counts()
    ent = entropy(counts)
    return(ent)

x1 = [0, 0, 1]
x2 = [0, 1, 0]



y = [0, 1, 1]



E = Ent(y)



LR = 0.1
#Initial Guess
W1 = 0.30
W2 = 0.45
TargetEnt = 0.4



CrossEnt = TargetEnt
W11=[]
W22=[]



#for e in range(0, 10):
e=0
CrossE=[]
while(CrossEnt >=TargetEnt):
    print("\nEpoch No.: ",e)

    yp = [] # Creation of blank list



    MSE = 0
    Cross_w1 = 0
    Cross_w2 = 0
    for r in range(0,len(x1)): #len(x) = 3
        temp = W1*x1[r] + W2*x2[r]
        t = 1/(1+math.exp(-temp)) #Sigmoid Activation
        yp.append(t)
        Cross_w1 = Cross_w1 + x1[r]*(t-y[r])
        Cross_w2 = Cross_w2 + x2[r]*(t-y[r])

        Ep = Ent(yp)
        CrossEnt = abs(E-Ep)
        print("E = %.3f\tEp = %.3f\t%.3f"%(E,Ep, abs(E-Ep)))
        CrossE.append(CrossEnt)



W1 = W1 - LR*Cross_w1
W2 = W2 - LR*Cross_w2
W11.append(W1)
W22.append(W2)
print("W1 = %3.3f\tW2 = %3.3f"%(W1,W2))
e=e+1



plt.subplot(1,3,1)
plt.plot(W11)
plt.xlabel("--- Epochs --->")
plt.ylabel("--- W1 --->")
plt.title("b0 Plot")
plt.grid()



plt.subplot(1,3,2)
plt.plot(W22)
plt.xlabel("--- Epochs --->")
plt.ylabel("--- W2 --->")
plt.title("b1 Plot")
plt.grid()



plt.subplot(1,3,3)
plt.plot(CrossE)
plt.xlabel("--- Epochs --->")
plt.ylabel("--- Cross Entropy --->")
plt.grid()
plt.show()
```
